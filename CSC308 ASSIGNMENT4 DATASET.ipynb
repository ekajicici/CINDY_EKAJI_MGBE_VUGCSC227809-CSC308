```python
from sklearn.model_selection import train_test_split
```

```python
# Applying Machine Learning (Classification — Predicting Customer Churn Scorre)
# Define X and y
X = df.drop('Churn Score', axis=1)
y = df['Churn Score'].map({'Yes': 1, 'No': 0})  # Convert churn to binary
```

```python
# Droping rows where y is NaN
df = df.dropna(subset=['Churn Score'])

# Then reassigning X and y
X = df.drop('Churn Score', axis=1)
y = df['Churn Score']
```

```python
# Binning Churn Score
def categorize_churn(score):
    if score <= 40:
        return 'Low'
    elif score <= 70:
        return 'Medium'
    else:
        return 'High'

y = df['Churn Score'].apply(categorize_churn)

# Train-testing split (70/30)
X = df.drop(columns=['Churn Score'])  # or select your features
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Training Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import (
    accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
)
```

```python
from sklearn.metrics import classification_report
# Making Predictions and Evaluate
y_pred = model.predict(X_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, zero_division=1))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, precision_recall_curve
```

```python
from sklearn.preprocessing import label_binarize
from sklearn.metrics import auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import RocCurveDisplay
import numpy as np
```

```python
# Binarizing the labels
y_test_bin = label_binarize(y_test, classes=['Low', 'Medium', 'High'])
y_pred_bin = label_binarize(y_pred, classes=['Low', 'Medium', 'High'])

# Computing ROC curve and ROC area for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(3):  # 3 classes
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_pred_bin[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Ploting ROC curve for each class
plt.figure(figsize=(6, 6))

for i in range(3):
    plt.plot(fpr[i], tpr[i], label=f'Class {["Low", "Medium", "High"][i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random classifier
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend(loc='best')
plt.show()
```
